```python
import matplotlib.pyplot as plt
import numpy as np
```

# Bai 1

```python
# make data
months=np.arange(1,13)
products= np.random.randint(1000,5000,size=len(months))
plt.figure(figsize=(9,5))
plt.plot(months,products,color='green',ls='--')
plt.title('Products by months',fontdict={'size':18,'weight':'bold','color':'blue'})
plt.xlabel('Months',fontdict={'weight':'bold'})
plt.ylabel('Products',fontdict={'weight':'bold'})
plt.grid(True,ls=':')
plt.xticks(range(1,13,1))
plt.tight_layout()
plt.text(x=1,y=3500,s='PS39392')
plt.show()
```

```python
#make data
quarter=np.arange(1,5)
revenue=np.random.randint(1000,5000,size=len(quarter))
ax=plt.bar(quarter,revenue,color='gray')
plt.title('Revenue over quarter',fontdict={'size':18,'weight':'bold','color':'orange'})
plt.xlabel('Quarter',fontdict={'weight':'bold'})
plt.ylabel('Revnue',fontdict={'weight':'bold'})
plt.xticks(range(1,5))
plt.grid(True,ls=':')
plt.text(x=1,y=3500,s='PS39392')
plt.tight_layout()
plt.show()
```

# Bai 3

```python
#make data
cate=['Shirt','Pant','Shoe','Acessories']
sales=np.random.randint(10000,50000,size=len(cate))
plt.figure(figsize=(8,8))
plt.pie(sales,explode=[0.01,0.01,0.01,0.01],labels=cate,colors=['lightblue','blue','green','gray'])
plt.title('Percent sales on each category',fontdict={'size':18},loc='left')
plt.axis('equal')
plt.show()
#ps39392
```

# Bai 4

```python
#make date
prices=np.random.uniform(50,300,size=100)
sales=np.random.uniform(2000,5000,size=100)+100*prices
plt.figure(figsize=(8,6))
plt.scatter(prices,sales)
plt.xlabel('prices')
plt.ylabel('sales')
plt.grid(True,ls=':')
plt.title('Coleration from Sales and Prices')
plt.text(x=50,y=15000,s='PS39392')
plt.tight_layout()
plt.show()
```

```python
# make data
months=np.arange(1,13)
products= np.random.randint(1000,5000,size=len(months))
sales=np.random.randint(1000,5000,size=len(months))
# plt.figure(figsize=(9,5))
plt.bar(months,products,color='gray')
plt.xlabel('Months',fontdict={'weight':'bold'})
plt.ylabel('Products',fontdict={'weight':'bold'})
plt.twinx()
plt.plot(months,sales,color='orange',ls='--',marker='o')
plt.ylabel('Sales',fontdict={'weight':'bold'})
plt.title('Products and sales over month',fontdict={'size':18,'weight':'bold','color':'lightblue'})
plt.grid(True,ls=':')
plt.xticks(range(1,13,1))
plt.tight_layout()
plt.text(x=1,y=3500,s='PS39392')
plt.show()
```
